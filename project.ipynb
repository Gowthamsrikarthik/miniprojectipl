```python
import pandas as pd
filepath="Match.csv"
def readcsv(filepath):
    return pd.read_csv(filepath)
df=readcsv(filepath)
df
```

```python
l=list(df["Season_Year"])
def count(l):
    for i in range(2008,2018):
        print("year",i,"no of matches:",l.count(i))
    return
count(l)
```

```python
l=list(df["Season_Year"])
li=list(df['match_winner'])
ck=[]
for i in range(0,637):
    if(l[i]==2011):
        ck.append(li[i])
csk=ck.count('Chennai Super Kings')
rr=ck.count('Rajasthan Royals')
rcb=ck.count('Royal Challengers Bangalore')
mi=ck.count('Mumbai Indians')
pw=ck.count('Pune Warriors')
kkr=ck.count('Kolkata Knight Riders')
kxp=ck.count('Kings XI Punjab')
dc=ck.count('Deccan Chargers')
ktk=ck.count('Kochi Tuskers Kerala')
dd=ck.count('Delhi Daredevils')
```

```python
from matplotlib.pyplot import *
divisions=['C S K', 'R R','R C B','M I','P W','K K R','K X P','D C','K T K','D D']
matchwins=[csk,rr,rcb,mi,pw,kkr,kxp,dc,ktk,dd]
bar(divisions,matchwins,color='yellow',width=0.40)
```
